```python
import matplotlib.pyplot as plt

sales_by_category=df.groupby("Product_Category")["Sales_Amount"].sum().sort_values(ascending=False).head(10)
explode=(0.2,0,0,0,0,00,0,0,0,0)
plt.figure(figsize=(10,5))
plt.pie(sales_by_category,labels=sales_by_category.index,explode=explode,autopct="%1.1f%%")
plt.title("Sales by Product Category")
```

```python
import pandas as pd
import numpy as np
import csv
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Laptop", "Mobile",
                "Headphones", "Tablet", "Laptop", "Mobile", "Tablet"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics",
                 "Accessories", "Electronics", "Electronics", "Electronics", "Electronics"],
    "Sales": [80000, 50000, 30000, np.nan, 45000, 7000, np.nan, 85000, 52000, 31000],
    "Quantity": [5, 10, 8, 7, np.nan, 15, 9, 6, 12, np.nan]
}

df = pd.DataFrame(data)

# Save to CSV
file_path = "sales2.csv"
df.to_csv(file_path, index=False)

df
```

```python
df["Quantity"].fillna(df["Quantity"].mean(),inplace=True)
df["Sales"].fillna(df["Sales"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
avg_sales = df["Sales"].mean()
avg_quantity = df["Quantity"].mean()

print("\nAverage Sales:", avg_sales)
print("Average Quantity:", avg_quantity)
```

```python

avg_sales_by_product = df.groupby("Product")["Sales"].mean()
avg_sales_by_category = df.groupby("Category")["Sales"].mean()

print("\nAverage Sales by Product:")
print(avg_sales_by_product)
print("\nAverage Sales by Category:")
print(avg_sales_by_category)
```
